First, we will have a very simple example using coupling blocks to perform the generation of the make moons dataset

```python
# lets start by generating our dataset, which can be easily done using sklear
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

# Parameters
n_samples = 5000  # Number of data points
noise = 0.1      # Noise level of the data

# Generate the dataset
X, y = make_moons(n_samples=n_samples, noise=noise, random_state=42)

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.title('Make Moons Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
# Now lets import them to keras tensors
import tensorflow as tf
from tensorflow import keras
import tensorflow_probability as tfp
tfd = tfp.distributions
import numpy as np

# Convert the data to tensors
X_tensor = tf.convert_to_tensor(X, dtype=tf.float32)
y_tensor = tf.convert_to_tensor(y, dtype=tf.int32)
```

```python
# Now we have also to create the base distirbution to allows us to calculate the log likelihood
def create_2d_gaussian(=0, =1.0):
    return tfd.MultivariateNormalDiag(
        loc=[, ],
        scale_diag=[, ]
    )

# Lets plot our base distribution to check its behaviour
#import matplotlib.pyplot as plt
#%matplotlib inline

gaussian_2d_base_dist = create_2d_gaussian()
gaussian_2d_samples = gaussian_2d_base_dist.sample(5000)
plt.scatter(gaussian_2d_samples[:, 0], gaussian_2d_samples[:, 1])
plt.title("Base Distribution")
plt.show()


# little test, gaussian of the point of the 2d gaussian - this is just a test, exclude this later!!!
probability_density = gaussian_2d_base_dist.prob([X_tensor])
print( np.shape(probability_density) )
plt.hist(probability_density[0, :])
plt.title("Base Distribution - make moons points")
plt.show()
print(probability_density[:, 0])
```

```python
# Now we can start setting up the normalizing flows. For this exercise we will use the affine coupling
# Which consists only of a scale and shift operation

# This should receive the tensors containing the parameters for the gamma, betta and the inputs x 
def affine_coupling(x, inputs):
    gamma = tf.math.exp(x[:,:2])
    beta  = x[:,2:]
    #print(np.shape(gamma))
    #print(np.shape(beta))
    return gamma*inputs + beta, gamma 

# We need to calculate the jacobian of the transformation in order to calculate the loss function
# the jacobian is simply the derivative w.r.t x!
def affine_jacobian(gamma, beta, x):
    return gamma

def flow_loss(x, inputs):
    
    # now that we have the outputs of the network (parameters of the affine transformation) we
    outputs, gammas =  affine_coupling(x, inputs)
    
    #print(x.shape)
    return tf.reduce_mean(-tf.math.log( gaussian_2d_base_dist.prob([outputs])) + tf.math.log(abs(1/gammas[:,0])) + tf.math.log(abs(1/gammas[:,1]))  ) # Lets not use the jacobian term for now!

def apply_flow(x, inputs):
    
    # now that we have the outputs of the network (parameters of the affine transformation) we
    outputs, gammas =  affine_coupling(x, inputs)
    return outputs

# The parameters gamma and beta are estimated using neural networks. Lets set a neural network for that
# how can we perform this? What should be the input and output dimension of the network?
# Lets try four? Two gammas and two betas?

# Defining our model

# Model building
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(2,)),  # Input layer with 2 features
    keras.layers.Dense(64, activation='relu'),  # Hidden layer with 10 neurons
    keras.layers.Dense(4, activation='sigmoid')  # Output layer with 1 neuron
])


## I think it would be better to create a class here, no?
optimizer = tf.keras.optimizers.Adam()
```

# In the first exercise we will only use one single transformation

```python
# Training Loop
# Need to implement the batch training!
epochs = 10
def train_model(model, X_train, y_train, epochs):
    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            predictions = model(X_train, training=True)
            loss = flow_loss( predictions, X_train)
        
        gradients = tape.gradient(loss, model.trainable_variables)
        optimizer.apply_gradients(zip(gradients, model.trainable_variables))
        
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss.numpy()}")
```

```python
# Train the model
train_model(model, X_tensor, y_tensor, epochs=500)
```
